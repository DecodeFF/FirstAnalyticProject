```python
df = pd.merge(data_equipment, data_equipment_personnel, on='day', how='inner')
df.head()
```

```python
df=df.drop("personnel*",axis=1)
df=df.drop("date_y",axis=1)
df.rename(columns={'date_x': 'date'}, inplace=True)
```

```python
df.info()
```

```python
plt.figure(figsize=(8, 7))

plt.plot(df['aircraft'], color='red', linewidth=1,label='Aircraft')
plt.plot(df['helicopter'], color='blue', linewidth=1,label='Helicopter')
plt.plot(df['MRL'], color='purple', linewidth=1,label='MRL')
plt.plot(df['naval ship'], color='pink', linewidth=1,label='Naval Ship')
plt.plot(df['anti-aircraft warfare'], color='green', linewidth=1,label='Anti-aircraft warfare')
plt.plot(df['special equipment'], color='black', linewidth=1,label='Special Equipment')


plt.xlabel('Дні')
plt.ylabel('Кількість')
plt.title('Знищено')
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8, 7))

plt.plot(df['tank'], color='green', linewidth=1,label='Tank')
plt.plot(df['field artillery'], color='purple', linewidth=1,label='Artillery')
plt.plot(df['drone'], color='blue', linewidth=1,label='Drone')
#plt.plot(df['personnel'], color='orange', linewidth=1,label='Personnel')

plt.xlabel('Дні')
plt.ylabel('Кількість')
plt.title('Знищено')
plt.grid(True)
plt.legend()

plt.show()
```

```python
differences = [df["personnel"].iloc[i+1] - df["personnel"].iloc[i] for i in range(len(df["personnel"])-1)]
max_increase_index = differences.index(max(differences))
max_increase = max(differences)

max_increase_day = df["day"].iloc[max_increase_index + 1]

print("Найбільший приріст знищеної живої сили це - "+ str(max_increase)+' це було',str(max_increase_day)+'-го дня')

```

```python
plt.figure(figsize=(8, 7))


plt.plot(df['personnel'], color='orange', linewidth=1,label='Personnel')

plt.xlabel('Дні')
plt.ylabel('Кількість')
plt.title('Знищено особового складу')
plt.grid(True)
plt.legend(["Найбільший приріст живої сили це - "+ str(max_increase)+' це було '+str(max_increase_day)+'-го дня'])

plt.show()
```
